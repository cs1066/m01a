### Push Today's Files to Your GitHub Repo

Right now, today's files exist in the codespace's filesystem. If you were to delete this codespace, these files would disappear with it. To make sure you don't lose these files, you'll want to turn the color of these new files' filenames from green with a "U" to their right into black (with no letter to the right). When I say this, I'm looking at the IDE's file explorer panel.

> In the terminal, executing `git status` gives you similar information.

Pushing these new files to your GitHub repo will ensure that you don't lose them. This is a multistep process:

1.  You "stage changes" either by clicking the plus-sign (+) next to each filename in the Source Control panel, or by running `git add <filename>` in the terminal window.

2.  With the changes you want staged, you then put a message in the textbox above the blue "Commit" button and click the "Commit" button (when working in the Source Control panel), or by running `git commit -m "<commit_message>"` in the terminal window. This commits your new files to the codespace's copy of your GitHub repo, which unfortunately will still disappear if you delete the codespace.

> NOTE: If you leave the `-m` flag off the git-commit command, git will open an editor and allow you to type a commit message. Including a commit message is a Good Practice. Ask an AI chatbot how to change the default editor into one that you know how to use.

3.  So, the last step is to push what you've committed to your GitHub repo. You can either click the "Push" icon above the list of commit messages, or you can run `git push` in the terminal window. Your work now appears in your CS1066 GitHub repo, and the files will appear in any new codespace you create (because codespaces pull from your GitHub repo).

Back in the IDE's file browser panel, the filenames of the new files should be black, and there should be no letter to the right.

> In the terminal, running `git status` now should report that "Your branch is up to date with 'origin/main'" and that there's "nothing to commit, working tree clean."

Check `https://github.com/<your_username>/cs1066` now and you'll see that the new files are there.

### Historical Categorization of Coding Environments

*   **No-code environments** expect no coding knowledge. They provide visual interfaces or ask for some configuration information. They typically target common, straightforward business tasks. If your problem fits their solution, you’re good. If not, you’re out of luck.

*   **Low-code environments** are like no-code environments, but they provide you with the option to write code in support of some unusual functionality. They are a step toward flexibility.

*   **High-code environments** offer maximum flexibility, but they required you to be an expert programmer.

With the integration of AI into IDEs (like the one you're using), you can follow any of these three approaches.

We'll work in an IDE so that we can dive deep when necessary. But most of the time, we'll describe what we want in English (not a computer programming language), and AI will produce runnable code in computer programming languages we don't fully understand.

### The Python Programming Language

Mostly, our target programming language will be Python, which is one of the world's most popular programming language.

It is a good choice because:

1.  Python is popular. Python offers a huge range of libraries that make it easy to build tools for many different application domains. With AI, you don't have to **discover** or **fully understand** these many libraries.
2.  Python is fairly easy to read. With a bit of practice, you can understand the gist of what some AI-produced code is asking a computer to do. In other words, with a small bit of knowledge about how to read Python, you can check the AI-generated code.

### Reading vs. Writing Code

Even software engineers spend much more time reading code than writing it.

Learning to read Python is much easier than learning to write it. You may decide that you want to learn to write code for yourself, but in this class, I'll **never ask you to write code without the aid of an AI**.

But I will ask (and teach) you to **read code for its gist**. Why? Think about your life.

*   When you want a task done, are you more comfortable when you have an idea of what’s involved?
*   With a little bit of knowledge, is it easier to pick someone who will do it well?
*   Is it easier to spot problems in their work (before you pay them for the work)?

AIs are not trustworthy. To be able to depend on what AIs code, you have to be able to check this code. To be able to check the code, you need to be able to read, understand, and evaluate code.
